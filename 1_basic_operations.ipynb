### Column Filtering

```python
df_trimmed = df.select("title","imdb_rating")
display(df_trimmed)
df_trimmed.count()
df_trimmed.printSchema()
```

### Row Filtering

```python
df_hollywood_2000_2010 = df.filter(
    (df["industry"] == "Hollywood") &
    (df["release_year"] >= 2000) &
    (df["release_year"] <= 2010)
)
display(df_hollywood_2000_2010)
```

```python
unique_industries = df.select("industry").distinct()
display(unique_industries)
```

```python
df = df.withColumn("profit", df["revenue"] - df["budget"])
display(df)
```
