### Parent Incremental Data Load

- there is no pipileine for getting this data in this sample project
- so compiled an aggregated data file and uploaded it into a Volume
- then copy that data into the Gold table 

```python
%sql

COPY INTO atlikon.gold.fact_orders
FROM (
  SELECT 
    CAST(date AS DATE) AS date,
    product_code,
    customer_code,
    CAST(sold_quantity AS BIGINT) AS sold_quantity
  FROM '/Volumes/atlikon/gold/parent_incremental_upload/*.csv'
)
FILEFORMAT = CSV
FORMAT_OPTIONS ('header' = 'true');
```
